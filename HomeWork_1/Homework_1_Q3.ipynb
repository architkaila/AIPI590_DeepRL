```python
## Library Imports
import gym
import numpy as np
from stable_baselines3 import A2C
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.evaluation import evaluate_policy
```

### Creating GYM Environments For Training & Evaluation 

```python
## Training environment of Cartpole
train_env = gym.make('CartPole-v1')

## Evalutation environment of Cartpole
evaluation_env = gym.make('CartPole-v1')
```

### Training Using "n-Step Advantage" 
- set use_n_step_advantage=True
- use default gae_lambda value

```python
## Initialize an Agent (Actor Critic)
model_n_step = A2C('MlpPolicy', train_env, verbose=1, tensorboard_log="./a2c_cartpole_tensorboard/", use_n_step_advantage=True)

## Training the agent for 100000 steps
model_n_step.learn(total_timesteps=100000)
```

```python
## Evaluating the agent using 100 episodes
reward_mean, reward_std = evaluate_policy(model_n_step, evaluation_env, n_eval_episodes=100)
print(f'Mean Reward: {np.round(reward_mean,2)} +/- {np.round(reward_std,2)}')
```

### Training With "Turned Off GAE"
- set use_n_step_advantage=False
- set gae_lambda = 0 (pass in A2C object initializer)

```python
## Initialize an Agent (Actor Critic)
model_no_gae = A2C('MlpPolicy', train_env, verbose=1, gae_lambda=0, tensorboard_log="./a2c_cartpole_tensorboard/", use_n_step_advantage=False)

## Training the agent for 100000 steps
model_no_gae.learn(total_timesteps=100000)
```

```python
## Evaluating the agent using 100 episodes
reward_mean, reward_std = evaluate_policy(model_no_gae, evaluation_env, n_eval_episodes=100)
print(f'Mean Reward: {np.round(reward_mean,2)} +/- {np.round(reward_std,2)}')
```

### Training With "Default GAE Advantage"
- set use_n_step_advantage=False
- use default gae_lambda value (Using original GAE with default params)

```python
## Initialize an Agent (Actor Critic)
model_default = A2C('MlpPolicy', train_env, verbose=1, tensorboard_log="./a2c_cartpole_tensorboard/", use_n_step_advantage=False)

## Training the agent for 100000 steps
model_default.learn(total_timesteps=100000)
```

```python
## Evaluating the agent using 100 episodes
reward_mean, reward_std = evaluate_policy(model_default, evaluation_env, n_eval_episodes=100)
print(f'Mean Reward: {np.round(reward_mean,2)} +/- {np.round(reward_std,2)}')
```
