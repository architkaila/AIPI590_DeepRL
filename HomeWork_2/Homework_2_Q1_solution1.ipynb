### Importing Libraries

```python
## Library Imports
import gym
import numpy as np

from stable_baselines3 import DQN
from stable_baselines3.common.evaluation import evaluate_policy
```

### Creating GYM Environments For Training & Evaluation 

```python
## Training environment of Cartpole
train_env = gym.make('CartPole-v1')

## Evalutation environment of Cartpole
evaluation_env = gym.make('CartPole-v1')
```

### Training "Vanilla DQN" 
- use default enable_double_dqn value = False

```python
## Initialize a Vanilla DQN
dqn_vanilla = DQN('MlpPolicy', 'CartPole-v1', verbose=1)

# Random agent, before training
reward_mean, reward_std = evaluate_policy(dqn_vanilla, evaluation_env, n_eval_episodes=100, deterministic=True)
print(f'Random DQN Mean Reward: {np.round(reward_mean,2)} +/- {np.round(reward_std,2)}')
```

```python
## Training the Vanilla DQN for 150000 steps
dqn_vanilla.learn(total_timesteps=150000)
```

```python
## Evaluating the Vanilla DQN using 100 episodes (After Training)
reward_mean, reward_std = evaluate_policy(dqn_vanilla, evaluation_env, n_eval_episodes=100, deterministic=True)
print(f'Trained DQN (Vanilla) Mean Reward: {np.round(reward_mean,2)} +/- {np.round(reward_std,2)}')
```

### Training "Double DQN" 
- set enable_double_dqn value = True

```python
## Initialize a Double DQN
dqn_double = DQN('MlpPolicy', 'CartPole-v1', verbose=1, enable_double_dqn=True)

## Random Double DQN, before training
reward_mean, reward_std = evaluate_policy(dqn_double, evaluation_env, n_eval_episodes=100, deterministic=True)
print(f'Random DQN Mean Reward: {np.round(reward_mean,2)} +/- {np.round(reward_std,2)}')
```

```python
## Training the Double DQN for 100000 steps
dqn_double.learn(total_timesteps=150000)
```

```python
## Evaluating the Double DQN using 100 episodes (After Training)
reward_mean, reward_std = evaluate_policy(dqn_double, evaluation_env, n_eval_episodes=100, deterministic=True)
print(f'Trained Double DQN Mean Reward: {np.round(reward_mean,2)} +/- {np.round(reward_std,2)}')
```

## Final Results (Post Running)
